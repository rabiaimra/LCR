### Classification using KNN

Let's set up our workspace and use the **Wine dataset** from `scikit-learn`. This dataset contains 178 wine samples with 13 chemical features, used to classify wines into different classes based on their origin.

The **response variable** is `class`, which indicates the type of wine. We'll use all of the chemical features to predict this response variable.

```python
# Import standard libraries
import pandas as pd
import numpy as np
import random
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import recall_score, precision_score
from sklearn.model_selection import cross_validate
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score
```

```python
from sklearn.datasets import load_wine

# Load the Wine dataset
wine_data = load_wine()

# Convert to DataFrame
wine_df = pd.DataFrame(wine_data.data, columns=wine_data.feature_names)

# Bind the 'class' (wine target) to the DataFrame
wine_df['class'] = wine_data.target

# Display the DataFrame
wine_df
```

#### **Question 2:** 
#### Standardization and data-splitting

Next, we must preform 'pre-processing' or 'data munging', to prepare our data for classification/prediction. For KNN, there are three essential steps. A first essential step is to 'standardize' the predictor variables. We can achieve this using the scaler method, provided as follows:

```python
# Select predictors (excluding the last column)
predictors = wine_df.iloc[:, :-1]

# Standardize the predictors
scaler = StandardScaler()
predictors_standardized = pd.DataFrame(scaler.fit_transform(predictors), columns=predictors.columns)

# Display the head of the standardized predictors
print(predictors_standardized.head())
```

(i) Why is it important to standardize the predictor variables?

Standardizing the predictor variables ensures that they are on a comparable scale. This is important because KNN relies on distance metrics to identify the nearest neighbors. If the predictor variables are on different scales, those with larger ranges can disproportionately influence the distance calculations, leading to misleading results and estimations. Standardization ensures that each variable has an equal impact on the model by contributing equally to the distance metric. 

(ii) Why did we elect not to standard our response variable `Class`?

We do not standardize the response variable Class ecause it is a categorical variable. Standardization is a transformation applied to continuous numerical variables, and applying it to categorical values would distort their meaning. Besides, in KNN, the response variable is not used in distance calculations, so there is no need to standardize it. 

(iii) A second essential step is to set a random seed. Do so below (Hint: use the random.seed function). Why is setting a seed important? Is the particular seed value important? Why or why not?

np.random.seed(1)

Setting a random seed is important because it ensures reproducibility. When randomness is involved (such as splitting data, sampling, or generating random numbers) the results can change each time the code runs. By setting a seed, we ensure the same random behavior occurs each time, which is crucial for consistent and repeatable results in testing or comparison. 

The particular seed value does not matter. You can use any integer value for seed(). So, instead of np.random.seed(1), I could say np.random.seed(153). What matters is that the same seed must be used to reproduce that specific random result, allowing your future self or others to replicate your results exactly. 

(iv) A third essential step is to split our standardized data into separate training and testing sets. We will split into 75% training and 25% testing. The provided code randomly partitions our data, and creates linked training sets for the predictors and response variables. 

Extend the code to create a non-overlapping test set for the predictors and response variables.

```python
# set a seed for reproducibility
np.random.seed(123)

# split the data into a training and testing set. hint: use train_test_split !

# Your code here ...

y = wine_df['class']

# Split the data into training and test sets 
X_train, X_test, y_train, y_test = train_test_split(
    predictors_standardized, y,
    train_size=0.75,
    stratify=y, 
    random_state=123
)

# Print the shape of the splits
print(X_train.shape, X_test.shape)
print(y_train.shape, y_test.shape)

# Check the class distribution
print('Training set:\n', y_train.value_counts(normalize=True))
print('Actual data: \n', y.value_counts(normalize=True))
```

#### **Question 3:**
#### Model initialization and cross-validation
We are finally set to fit the KNN model. 


Perform a grid search to tune the `n_neighbors` hyperparameter using 10-fold cross-validation. Follow these steps:

1. Initialize the KNN classifier using `KNeighborsClassifier()`.
2. Define a parameter grid for `n_neighbors` ranging from 1 to 50.
3. Implement a grid search using `GridSearchCV` with 10-fold cross-validation to find the optimal number of neighbors.
4. After fitting the model on the training data, identify and return the best value for `n_neighbors` based on the grid search results.

```python
# Your code here...

# Initialize the KNN classifier
knn = KNeighborsClassifier()

# Defining a parameter grid for n_neighbors ranging from 1 to 50
parameter_grid = {
    "n_neighbors": range(1, 51)
}

# Implementing a grid search with 10-fold cross-validation
wine_tune_grid = GridSearchCV(
    estimator=knn,
    param_grid=parameter_grid,
    cv=10
)

# Fitting the model on the training data
wine_tune_grid.fit(
    X_train,
    y_train
)

# Accuracies grid
accuracies_grid = pd.DataFrame(wine_tune_grid.cv_results_)
print(accuracies_grid)

# Print the no of neighbors with the highest accuracy
best_params = wine_tune_grid.best_params_
print("Best number of neighbors:", best_params)

```

#### **Question 4:**
#### Model evaluation

Using the best value for `n_neighbors`, fit a KNN model on the training data and evaluate its performance on the test set using `accuracy_score`.

```python
# Your code here...

print(type(best_params))

# Initialize KNN with the best number of neighbors
knn_best = KNeighborsClassifier(n_neighbors=best_params['n_neighbors'])

# Fit the model on training data
knn_best.fit(X_train, y_train)

# Make predictions on the test set
y_pred = knn_best.predict(X_test)

# Calculate accuracy on the test set
test_accuracy = accuracy_score(y_test, y_pred)

print("Test set accuracy is:", test_accuracy)
```
